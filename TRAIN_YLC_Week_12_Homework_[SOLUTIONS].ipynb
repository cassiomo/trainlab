**Run the code cell below to create the DataFrame.**

```python
df = pd.DataFrame({'River': ['Nile', 'Amazon', 'Yangtze', 'Mississippi', 'Yenisei'],
                   'Continent': ['Africa', 'South America', 'Asia', 'North America', 'Asia'],
                   'Length (km)': [6650, 6575, 6300, 6275, 5540],
                   'Countries': ['Egypt, Sudan, South Sudan, Ethiopia, Uganda, Congo, Kenya, Tanzania, Rwanda, Burundi',
                                 'Brazil, Peru, Bolivia, Colombia, Ecuador, Venezuela, Guyana',
                                 'China',
                                 'United States',
                                 'Russia']})
```

 Determine the names of the columns in this dataset.

#### **Problem #1.3**

Find out the length of the longest river in the dataset.

##### **Solution**

```python
df['Length (km)'].max()
```

#### **Problem #1.4**

List all the continents included in this dataset.

##### **Solution**

```python
df['Continent'].unique()
```

#### **Problem #1.5**

Calculate the average length of the rivers in the dataset.

##### **Solution**

#### **Problem #1.6**

Determine which countries are included in this dataset.

##### **Solution**

```python
df['Countries'].unique()
```

#### **Problem #1.7**

**Independently**, determine the total length of all rivers in this dataset combined.

##### **Solution**

```python
df['Length (km)'].sum()
```

#### **Problem #1.8**

Create a visualization with `River` on the x-axis and `Length (km)` on the y-axis. It is up to you to decide which visualization to create. Include a meaningful title and axes labels.



##### **Solution**

```python
plt.bar(df['River'], df['Length (km)'])

plt.title("Length in KM of Rivers")
plt.xlabel('River')
plt.ylabel('Length (km)')

plt.show()
```
